```python
import numpy as np
import math
import matplotlib.pyplot as mpl
```

```python
## Esta funcion genera una senal senoidal
def seno(ts ,a,f,theta):
    x=[]
    for t in ts:
        x.append(a*np.sin ((2*np.pi*f*t)+theta))
    return x

## Esta funcion genera una senal cosenoidal
def cos(ts ,a,f,theta):
    x=[]
    for t in ts:
        x.append(a*np.cos ((2*np.pi*f*t)+theta))
    return x

## Esta es la funcion principal del programa
def main(ini ,fin ,delta):
    ts=np.arange(ini ,fin ,delta)
    x1=seno(ts ,10 ,0.25 ,0)
    x2=seno(ts ,20 ,0.125 ,np.pi/2)
    mpl.plot(ts ,x1)
    mpl.plot(ts ,x2)
    mpl.grid()
    mpl.title('SENAL SENO CHIDA Y OTRA MAS CHIDA ')
    mpl.xlabel('Tiempo ')
    mpl.ylabel('Respuesta ')
    mpl.legend ((r'$10sen (\omega_1 t)$',r'$20sen (\omega_2 t+\pi/2)$'))
    mpl.savefig('ejem6.png',format='png',dpi =300)
    mpl.show()
```

```python
## Este es el punto de entrada al programa
if __name__ =="__main__":
    main (-5 ,15 ,0.1)
```
