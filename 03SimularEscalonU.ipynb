```python
import numpy as np
import matplotlib.pyplot as mpl
```

```python
## Esta funci√≥n simula un esalon 
def escalon(ts ,a,  k):
    x=[]
    for t in ts:
        if t<k:
            x.append (0)
        else:
            x.append(a)
    return x

## Esta es la funcion principal del programa
def main(ini ,fin ,delta):
    ts=np.arange(ini ,fin ,delta)
    x1=escalon(ts ,10 ,5)
    x2=escalon(ts ,1,0)
    mpl.plot(ts ,x1)
    mpl.plot(ts ,x2)
    mpl.grid()
    mpl.title('ESCALON NORMAL Y ESCALON CHIDO ')
    mpl.xlabel('Tiempo ')
    mpl.ylabel('Respuesta ')
    mpl.savefig('ejem2.png',format='png',dpi =300)
    mpl.show()
```

```python
## Este es el punto de entrada al programa
if __name__ =="__main__":
    main (-10 ,10 ,0.1)
```
